## HTTP Request para download das bacias relevantes delimitadas pela ONS

```python
url = "https://ons-dl-prod-opendata.s3.amazonaws.com/dataset/bacia_contorno/Bacias_Hidrograficas_SIN.zip"

# Download the zip file
with urllib.request.urlopen(url) as urlfile:
    data = urlfile.read()

# Extract the contents of the zip file to a folder
with zipfile.ZipFile(io.BytesIO(data)) as zip_file:
    zip_file.extractall("bacias_hidrograficas")
```

```python
# Download EAR_DIARIO_RESERVATORIOS_2021.csv
ear_url = 'https://ons-dl-prod-opendata.s3.amazonaws.com/dataset/ear_reservatorio_di/EAR_DIARIO_RESERVATORIOS_2021.csv'
ear_file = 'EAR_DIARIO_RESERVATORIOS_2021.csv'

response = requests.get(ear_url)

with open(ear_file, 'wb') as f:
    f.write(response.content)
    
# Download ENERGIA_VERTIDA_TURBINAVEL_2021.csv
cmo_url = 'https://ons-dl-prod-opendata.s3.amazonaws.com/dataset/cmo_se/CMO_SEMANAL_2021.csv'
cmo_file = 'CMO_SEMANAL_2021.csv'

response = requests.get(cmo_url)

with open(cmo_file, 'wb') as f:
    f.write(response.content)
    
# Download ENERGIA_VERTIDA_TURBINAVEL_2008.csv
cmo_url = 'https://ons-dl-prod-opendata.s3.amazonaws.com/dataset/cmo_se/CMO_SEMANAL_2008.csv'
cmo_file = 'CMO_SEMANAL_2008.csv'

response = requests.get(cmo_url)

with open(cmo_file, 'wb') as f:
    f.write(response.content)    
```

```python
df = pd.read_csv('AGRUPADO.csv')
cmo_21 = pd.read_csv('CMO_SEMANAL_2021.csv', delimiter=';')
cmo_08 = pd.read_csv('CMO_SEMANAL_2008.csv', delimiter=';')

cmo_08_mean = cmo_08.groupby('nom_subsistema')['val_cmomedia'].mean().reset_index()
cmo_08_mean['ano'] = 2008
cmo_21_mean = cmo_21.groupby('nom_subsistema')['val_cmomedia'].mean().reset_index()
cmo_21_mean['ano'] = 2021

cmo = pd.concat([cmo_08, cmo_21])

# Group by year and calculate the mean of "val_cmomedia" for each year
cmo = cmo.groupby(cmo['din_instante'].str[:4])['val_cmomedia'].mean().reset_index()

# Rename the columns
cmo.columns = ['ano', 'val_cmomedia']
# # Plot the data
# sns.scatterplot(data=cmo, x='ano', y='val_cmomedia')

df_filtered = df[df['classificação'].isin(['Antrópico', 'Não-Antrópico'])]

# group by 'ano'
df_grouped = df_filtered.groupby('ano')

# calculate the sum of 'area' for each group
sum_antr = df_grouped.apply(lambda x: x[x['classificação'] == 'Antrópico']['area'].sum())
sum_nao_antr = df_grouped.apply(lambda x: x[x['classificação'] == 'Não-Antrópico']['area'].sum())

# divide the sums and create a new DataFrame
df_preserv = pd.DataFrame({'ano': sum_antr.index, 'preserv': sum_antr / sum_nao_antr})

cmo['preserv'] = [0.467792, 0.460951]

sns.set_style("darkgrid")

# create scatter plot with regression line
sns.lmplot(x='val_cmomedia', y='preserv', data=cmo, height=7)

# set plot title and labels
plt.title('Scatter plot of CMO vs. preservation')
plt.xlabel('CMO')
plt.ylabel('Preservation')

# display plot
plt.show()
```
